```python
# Note: This is a comment in Python
# Python library imports
import requests # For HTTP requests
from PIL import Image # Image reading and manipulation
import io # For stream io
import numpy as np # Numerical Python library
import pandas as pd # Data manipulation library
import matplotlib.pyplot as plt # Plotting
# Tells Jupyter to plot within the notebook
%matplotlib inline
```

```python
# URL to a sample image
url = "https://raw.githubusercontent.com/harris-soh-copeland-puca/SampleFiles/master/bridge.jpg"

# HTTP GET request for an image from URL
response = requests.get(url)

# Check HTTP response is OK and throw error if not
response.raise_for_status()

# Get the image content
with io.BytesIO(response.content) as f:
    # Use PIL library to open image stream
    with Image.open(f) as img:
        # Display the image in Jupyer notebook
        plt.imshow(img)
```

**Split image into red, green, blue channels (three 2D arrays)**

```python
red, green, blue = img.split()
```

```python
# Convert image to numpy array from PIL Image
np_img = np.asarray(img)

# Set background to all zeros
tmp = np.zeros(np_img.shape, dtype='uint8')

# Layer on top the red channel
tmp[:,:,0] = red

# Plot
plt.imshow(tmp)
```

```python
# Convert image to numpy array from PIL Image
np_img = np.asarray(img)

# Set background to all zeros
tmp = np.zeros(np_img.shape, dtype='uint8')

# Layer on top the green channel
tmp[:,:,1] = green

# Plot
plt.imshow(tmp)
```

```python
# Convert image to numpy array from PIL Image
np_img = np.asarray(img)

# Set background to all zeros
tmp = np.zeros(np_img.shape, dtype='uint8')

# Layer on top the blue channel
tmp[:,:,2] = blue

# Plot
plt.imshow(tmp)
```
